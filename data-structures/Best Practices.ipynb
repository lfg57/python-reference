# Python Best Practices

## Ranges and Lists

### Looping over a range of numbers

**`ðŸ…´`**

**`# DON'T`**

````python
for i in [0, 1, 2, 3]:
    print(i**2, end=' ')    
````

**`# DO`**

````python
for i in range(4):
    print(i**2, end=' ')
````

### Looping over a collection

**`ðŸ…´`**

````python
colors = ['red', 'green', 'blue']
````

**`# DON'T`**

````python
for i in range(len(colors)):
    print(colors[i], end=' ')
````

**`# DO`**

````python
for color in colors:
    print(color, end=' ')
````

### Looping backwards

**`ðŸ…´`**

````python
colors = ['red', 'green', 'blue']
````

**`# DON'T`**

````python
for i in range(len(colors)-1, -1, -1):
    print(colors[i], end=' ')
````

**`# DO`**

````python
for color in colors[::-1]:
    print(color, end=' ')
````

````python
for color in reversed(colors):
    print(color, end=' ')
````

### Looping over a collection and indices

**`ðŸ…´`**

````python
colors = ['red', 'green', 'blue']
````

**`# DON'T`**

````python
for i in range(len(colors)):
    print("{} -> {}".format(i, colors[i]))
````

**`# DO`**

````python
for index,color in enumerate(colors):
    print("{} -> {}".format(index, color))
````

### Looping over two collections

**`ðŸ…´`**

````python
colors = ['red', 'green', 'blue']
things = ['car', 'trees', 'moon', 'note']
````

**`# DON'T`**

````python
n = min(len(colors), len(things))
for i in range(n):
    print("{:5} -> {}".format(colors[i], things[i]))
````

**`# DO`**

````python
for color,thing in zip(colors, things):
    print("{:5} -> {}".format(color, thing))
````

### Looping in sorted order

**`ðŸ…´`**

````python
colors = ['red', 'green', 'blue', 'amber', 'orange']
````

**`# DON'T`**

**`# DO`**

````python
for color in sorted(colors):
    print(color, end=' ')
````

````python
for color in sorted(colors, reverse=True):
    print(color, end=' ')
````

### Custom sort order

**`ðŸ…´`**

````python
colors = ['red', 'green', 'blue', 'amber', 'orange']
````

**`# DON'T`**

```python
# Deprecated in Python 3
def compare_len(c1, c2):
    if len(c1) < len(c2): return -1
    if len(c1) > len(c2): return 1
```

## Functions

### Call a function until a sentinel value

**`ðŸ…´`**

**`# DON'T`**

```python
blocks = []
while True:
    block = f.read(32)
    if block == '':
        break
    blocks.append(block)
```

**`# DO`**

```python
from functools import partial

blocks = []
for block in iter(partial(f.read, 32), ''):
    blocks.append(block)
```

### Distinguishing multiple exit points in loops

**`ðŸ…´`**

````python
seq = [2, 3, 1, 5, 0, 9, 12]
````

**`# DON'T`**

````python
def find_no(seq, target):
    found = False
    
    for index,value in enumerate(seq):
        if value == target:
            found = True
            break
    
    if not found:
        return -1
    
    return index
````

````python
find_no(seq, 9)
````

**`# DO`**

````python
def find(seq, target):
    for index,value in enumerate(seq):
        if value == target:
            return index
    else:
        return -1
````

````python
find(seq, 9)
````

## Dictionaries

### Looping over dictionary keys

**`ðŸ…´`**

````python
d = {'red':'car', 'green':'trees', 'blue':'moon'}
````

**`# DON'T`**

**`# DO`**

````python
for k in d:
    print(k, end=' ')
````

````python
for k in d.keys():
    if k.startswith('r'):
        print("{}:{}".format(k, d[k]), end='')
````

````python
dr = {k:d[k] for k in d if not k.startswith('r')}
dr
````

### Looping over dictionary keys and values

**`ðŸ…´`**

````python
d = {'red':'car', 'green':'trees', 'blue':'moon'}
````

**`# DON'T`**

**`# DO`**

````python
for k in d:
    print("{:5} -> {}".format(k, d[k], end=' '))
````

````python
for k in d.keys():
    if k.startswith('r'):
        print("{}:{}".format(k, d[k]), end='')
````

````python
dr = {k:d[k] for k in d if not k.startswith('r')}
dr
````
